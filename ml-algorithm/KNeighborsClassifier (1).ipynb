```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/valixonovai/ML/refs/heads/main/data_set/data.csv")
df
```

```python
df.describe(include="all")
```

```python
print("Dataset haqida qisqacha ma'lumotlar(dtype,column,index,row)")
print("-"*70)
df.info()
```

```python
print("B - Sog'lom \nM - Saraton")
df['diagnosis'].value_counts()
```

```python
# Diagnosis ustuni numeric qilib olamiz birinchi yo'li replace orqali yoki skleaningdagi LabelEncoding orqali
df['diagnosis'] = df['diagnosis'].replace({"M":1,"B":0})
df.head()
```

```python
# bizda Unnamed: 32 -- > ustun bor ekan uni oçhirib tashlaymiz ichi bosh ekan.
df.drop("Unnamed: 32", axis=1, inplace=True)
df.head()
```

```python
df.corrwith(df.diagnosis).abs().sort_values(ascending=False)
```

Birinchi biz datani target data va natijalarga ajaratib olsek \
Ikkinchi datani train va test ajratib olamiz.

```python
df.drop("id", axis=1, inplace=True)
X = df.drop("diagnosis", axis=1)
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=42, train_size=0.8)
```

```python
X_train.head()
```

```python
s_scaler = StandardScaler()
X_train_ss = s_scaler.fit_transform(X_train)
X_test_ss = s_scaler.fit_transform(X_test)
```

```python
X_train_ss[0]
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train_ss, y_train)
```

```python
predict_model = model.predict(X_test_ss)
print(classification_report(predict_model, y_test))
```

```python
print(confusion_matrix(predict_model, y_test))
```

Eng yaxshi `n_neighbors` qandey topamiz ?

```python
f = []
for i in range(1,25):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train_ss, y_train)
  p = knn.predict(X_test_ss)
  f.append(accuracy_score(p, y_test))
print("--"*32)
print(f"Eng yaxshi n_neighbors --> eng yaxshi qo'shilar soni bu {f.index(max(f))} ekan.")
print("--"*32)
plt.figure(figsize=(12,6))
plt.plot(range(1,25),f)
plt.grid()
plt.show()
```
